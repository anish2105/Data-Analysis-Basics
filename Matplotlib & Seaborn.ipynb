### MATPLOTLIB

Matplotlib is a Python 2D plotting library which produces publication quality figures in a variety of hardcopy formats and interactive environments across platforms. Matplotlib can be used in Python scripts, the Python and IPython shells, the Jupyter notebook, web application servers, and four graphical user interface toolkits.

Third party packages

A large number of third party packages extend and build on Matplotlib functionality, including several higher-level plotting interfaces (seaborn, holoviews, ggplot, ...), and two projection and mapping toolkits (basemap and cartopy).

matplotlib.pyplot is a collection of command style functions that make matplotlib work like MATLAB. Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.

In matplotlib.pyplot various states are preserved across function calls, so that it keeps track of things like the current figure and plotting area, and the plotting functions are directed to the current axes (please note that "axes" here and in most places in the documentation refers to the axes part of a figure and not the strict mathematical term for more than one axis).

Tip: In Jupyter Notebook, you can also include %matplotlib inline to display your plots inside your notebook.

# Plotting multiple Points

```python
#First create a numpy array and assign to x and y
x = np.array([2,4,6,8,10,12,14,16]) #This will be our x axis
y = x/2 #This will be our y axis

plt.figure(figsize = (10,5)) #.figure is used to set the total size of graph where in x-axis it is 10 units and in y axis its 5 units
plt.scatter(x,y,c = 'red') 
plt.show() #.show() is used to display the graph
```

```python
plt.plot([1,2,3,4])
plt.ylabel('vertical') #It gives name to the y axis
plt.xlabel('horizontal') #It gives name to the x axis
plt.show()
```

# Multiple plots with subplots

```python
name = ['John','Ellen','Meera','Raj','Rahul']
scores = [10 ,15,27,20,50]

plt.figure(figsize= (12,5))

plt.subplot(131)
#Here 131 means there is 1 row in which graphs are plotted
#3 columns and this subplot will be in position one
plt.bar(name,scores)

plt.subplot(132)
plt.scatter(name,scores)

plt.subplot(133)
plt.plot(name,scores)

plt.suptitle("Catorgorial Plotting") # This gives the name for Plot
plt.show()
```

```python
name = ['John','Ellen','Meera','Raj','Rahul']
scores = [10 ,15,27,20,50]

plt.figure(figsize= (12,5))

plt.subplot(221)
#Here 131 means there is 1 row in which graphs are plotted
#3 columns and this subplot will be in position one
plt.bar(name,scores)

plt.subplot(222)
plt.scatter(name,scores)

plt.subplot(223)
plt.plot(name,scores)

plt.suptitle("Catorgorial Plotting") # This gives the name for Plot
plt.show()
```

# SEABORN

Seaborn comes with a large number of high-level interfaces and customized themes that matplotlib lacks as it becomes difficult to figure out the settings that make plots attractive.

Mostly, matplotlib functions don’t work well with dataframes as seaborn does.

PS: Seaborn visualisations are based on matplotlib

### We are loading Pokemon.csv dataset to work with Seaborn

```python
import seaborn as sns
```

```python
my_data = pd.read_csv('Pokemon.csv')
```

```python
my_data.head(9)
```

```python
sns.lmplot(x = 'Attack' , y = 'Defense' , data = my_data) #Here lmplot() function is used to quickly plot the Linear Relationship between two(2) variables.
#lm for linear regression model
plt.plot()
```

```python
# To remove the remove the regession line use fit_reg = False
sns.lmplot(x = 'Attack' , y = 'Defense' , data = my_data , fit_reg = False)
```

```python
#We can use hue to highlight a particular values of column , here legendary pokemons are highlisted in orange
sns.lmplot(x = 'Attack' , y = 'Defense' , data = my_data , fit_reg = False , hue = 'Legendary')
```

```python
my_data['Legendary'].value_counts()
```

```python
sns.lmplot(x = 'Attack' , y = 'Defense' , data = my_data , fit_reg = False , hue = 'Generation')
```

Explicitly setting our y-limit and x-limit 

```python
fig = plt.figure()
a1 = fig.add_axes([0,0,1,1]) #[0,0,1,1] here 1,1 refers to width and height

x = np.arange(1,10)
a1.plot(x,np.exp(x),'r') #'r' is refers to color red
a1.set_title('range of numbers')

#Explicitly setting y-axis and x-axis value
plt.ylim(0,10000)
plt.xlim(0,15)

plt.show()
```

Box Plot is the visual representation of the depicting groups of numerical data through their quartiles. Boxplot is also used for detect the outlier in data set.

A box plot consist of 5 things.

Minimum
First Quartile or 25%
Median (Second Quartile) or 50%
Third Quartile or 75%
Maximum

```python
plt.figure(figsize = (15,15))
sns.boxplot(data = my_data)
```

```python
#Drop unwanted columns use drop function
my_data1 = my_data.drop(['#','Total','Generation','Legendary'],axis = 1)
my_data1
```

```python
plt.figure(figsize = (15,15))
sns.boxplot(data = my_data1)
```

```python
#To calculate correlation

corr = my_data1.corr()
corr
```

```python
#Creating a HeatMap for the correlation

plt.figure(figsize = (10,10))
sns.heatmap(corr)
#More the color becomes lighter , more the correlation and vice versa
```

```python
#To show the correlation number we use annot
plt.figure(figsize = (10,10))
sns.heatmap(corr,annot = True)
```

# Univariate Visualization

# Distplot

The most convenient way to take a quick look at a univariate distribution in seaborn is the distplot() function. 
By default, this will draw a histogram and fit a kernel density estimate (KDE). 
It is used basically for univariant set of observations and visualizes it through a histogram i.e. only one observation and hence we choose one particular column of the dataset.

```python
sns.distplot(my_data1['Defense'])
```

```python
sns.boxplot(my_data1['Defense'])
```

```python
#If we dont want a kde line then we can set it to false
sns.distplot(my_data1['Defense'],kde = False)
```

We can only plot just the kde line

```python
sns.kdeplot(my_data1['Defense'])
```

```python
#We can also shade the whole kde line
sns.kdeplot(my_data1['Defense'] , shade =True)
```

# Bivariate Distributions Visualisations

### Jointplot

```python
my_data.head()
```

```python
sns.jointplot(my_data['Attack'] , my_data['Defense'] )
```

```python
#To confirm their correlation
my_data[['Defense' , 'Attack']].corr()
```

```python
#To plot jointplot with kde ,use kind = kde inside jointplot
sns.jointplot(my_data['Attack'] , my_data['Defense'] , kind = 'kde')
plt.show()
```

```python
#For regression plot(i.e., lmplot) use kind = 'reg'
sns.jointplot(my_data['Attack'] , my_data['Defense'] , kind = 'reg')
```

We can also explicitly set the 'kind' of visualisation to be displayed

e.g: kind= “scatter” or “reg” or “resid” or “kde” or “hex”

## Visaulising more than two variables: Pairwise Bivariate Distributions-Using Pairplot()

```python
sns.pairplot(my_data[['Defense' , 'Attack' , 'HP']] , kind = 'scatter')
plt.show()
```

We can change the diagonal kind

```python
sns.pairplot(my_data[['Defense' , 'Attack' , 'HP']] , kind = 'scatter' , diag_kind = 'kde')
```
